```python
import re

# Uses HEADER_LINE you already detected:
# HEADER_LINE = '04/03/2026 The Penal Code, 1860'

def normalize_legal_text(page_text: str, header_line: str | None) -> str:
    if not page_text:
        return ""

    # Normalize line endings first
    text = page_text.replace("\r\n", "\n").replace("\r", "\n")

    # 1) Remove repeating header if present
    # Case A: header is its own line (most pages)
    if header_line:
        lines = text.splitlines()
        # find first non-empty line
        i = 0
        while i < len(lines) and not lines[i].strip():
            i += 1
        if i < len(lines) and lines[i].strip() == header_line:
            lines.pop(i)
        text = "\n".join(lines)

        # Case B: header is glued to the next token (your page 120 case)
        # Remove header as a prefix even if immediately followed by letters (e.g., "1860substance,")
        # Only do this at the very start of the page to avoid accidental deletions.
        text = re.sub(rf"^\s*{re.escape(header_line)}(?=\S)", "", text)

    # 2) Fix "stuck words" conservatively: insert space between lowercase -> Uppercase
    text = re.sub(r"([a-z])([A-Z])", r"\1 \2", text)

    # 3) Normalize internal whitespace (tabs/multiple spaces)
    text = re.sub(r"[ \t]+", " ", text)

    # 4) Normalize excessive blank lines
    text = re.sub(r"\n{3,}", "\n\n", text)

    # 5) Trim spaces around newlines
    text = re.sub(r" *\n *", "\n", text).strip()

    return text

def show_before_after(page_index: int, n: int = 700):
    raw = documents[page_index].page_content
    clean = normalize_legal_text(raw, HEADER_LINE)

    raw_first = raw.strip().splitlines()[0] if raw.strip() else ""
    clean_first = clean.strip().splitlines()[0] if clean.strip() else ""

    print(f"\n========== PAGE {page_index} ==========")
    print("RAW first line:  ", repr(raw_first))
    print("CLEAN first line:", repr(clean_first))

    print("\n--- RAW (first chars) ---")
    print(raw[:n])

    print("\n--- CLEAN (first chars) ---")
    print(clean[:n])

# Validate the cleaning on key pages
for p in [0, 5, 50, 120, 199]:
    show_before_after(p, n=700)

# Find Section 378 page (citation readiness test)
SECTION_TO_FIND = 378
section_re = re.compile(rf"(?m)^\s*{SECTION_TO_FIND}\.\s")  # start-of-line "378. "

hit_page = None
for i, doc in enumerate(documents):
    cleaned = normalize_legal_text(doc.page_content, HEADER_LINE)
    if section_re.search(cleaned):
        hit_page = i
        break

print("\n==============================")
print("First page containing section", SECTION_TO_FIND, ":", hit_page)

if hit_page is not None:
    cleaned = normalize_legal_text(documents[hit_page].page_content, HEADER_LINE)
    m = section_re.search(cleaned)
    start = max(0, m.start() - 300)
    end = min(len(cleaned), m.start() + 900)
    print("\n--- CLEANED CONTEXT AROUND MATCH ---")
    print(cleaned[start:end])
```

```python
import re

# 1) Build one full cleaned text with explicit page markers (critical for citations)
PAGE_MARKER = "\n\n<<<PAGE:{page}>>>\n\n"

clean_pages = []
for i, doc in enumerate(documents):
    clean_pages.append(PAGE_MARKER.format(page=i) + normalize_legal_text(doc.page_content, HEADER_LINE))

full_text = "\n".join(clean_pages)

print("Full cleaned text length:", len(full_text))

# 2) Section boundary patterns (handles: "Theft378." and "378. Theft" and "378.Theft")
# Explanation:
# - capture an optional heading text before the number (letters/spaces)
# - capture section number 1-4 digits
# - require a dot after the number
section_boundary = re.compile(
    r"(?m)(?P<heading>[A-Za-z][A-Za-z \-']{0,80}?)?\s*(?P<section>\d{1,4})\.\s*"
)

# 3) Find boundaries, but filter out obvious false matches like years 1860/1937/1961 in brackets.
# We'll only accept section numbers <= 600 (Penal Code max is 511), adjust later per Act.
MAX_REASONABLE_SECTION = 600

matches = []
for m in section_boundary.finditer(full_text):
    sec = int(m.group("section"))
    if sec > MAX_REASONABLE_SECTION:
        continue

    # Context filter: avoid matches inside [...] or (...) often used for amendment years/orders
    ctx_start = max(0, m.start() - 25)
    ctx = full_text[ctx_start:m.start() + 10]

    # If immediately preceded by a digit (like part of 1961), skip
    if m.start() > 0 and full_text[m.start()-1].isdigit():
        continue

    # If inside square brackets nearby, likely amendment note, skip
    if "[" in ctx and "]" not in ctx:
        # open bracket near match but not closed yet -> probably bracketed note
        continue

    heading = (m.group("heading") or "").strip()
    # Remove trailing artifacts that sometimes appear
    heading = re.sub(r"\s+", " ", heading)

    matches.append((m.start(), m.end(), sec, heading))

print("Candidate section boundaries found:", len(matches))
print("First 20 candidates (sec, heading):")
for item in matches[:20]:
    print(item[2], "|", item[3])

# 4) Build sections by slicing between boundaries
sections = {}  # sec_num -> {"heading":..., "text":..., "start":..., "end":...}
for idx, (s_start, s_end, sec, heading) in enumerate(matches):
    next_start = matches[idx + 1][0] if idx + 1 < len(matches) else len(full_text)
    sec_text = full_text[s_end:next_start].strip()

    # Basic cleanup: remove page markers from section body but keep page ids discoverable later
    # (We will compute page ranges in next step)
    sections[sec] = {
        "heading": heading,
        "text": sec_text,
        "start_offset": s_start,
        "end_offset": next_start,
    }

print("\nTotal unique sections parsed:", len(sections))
print("Does section 378 exist?", 378 in sections)

if 378 in sections:
    print("\n--- Section 378 heading ---")
    print(sections[378]["heading"])
    print("\n--- Section 378 preview ---")
    print(sections[378]["text"][:800])
```

```python
import re

def normalize_legal_text_v2(page_text: str, header_line: str | None) -> str:
    # start from your existing normalize_legal_text
    text = normalize_legal_text(page_text, header_line)

    # Remove common footer patterns from Bangladesh Code print PDFs
    # Example: "bdlaws.minlaw.gov.bd/act-print-11.html 137/200"
    text = re.sub(r"(?im)^bdlaws\.minlaw\.gov\.bd/act-print-\d+\.html\s+\d+/\d+\s*$", "", text)

    # Sometimes footer appears without start-of-line due to wrapping; remove if it appears as a standalone fragment
    text = re.sub(r"(?i)\bbdlaws\.minlaw\.gov\.bd/act-print-\d+\.html\b", "", text)

    # Remove trailing "X/Y" page counters if they appear alone on a line
    text = re.sub(r"(?m)^\s*\d{1,4}/\d{1,4}\s*$", "", text)

    # Normalize blank lines again after deletions
    text = re.sub(r"\n{3,}", "\n\n", text).strip()
    return text

# Rebuild full text with page markers using v2 cleaner
PAGE_MARKER = "\n\n<<<PAGE:{page}>>>\n\n"

clean_pages_v2 = []
for i, doc in enumerate(documents):
    clean_pages_v2.append(PAGE_MARKER.format(page=i) + normalize_legal_text_v2(doc.page_content, HEADER_LINE))

full_text_v2 = "\n".join(clean_pages_v2)

print("Full cleaned v2 text length:", len(full_text_v2))

# Improved boundary regex:
# - Require that the section number is followed by a dot
# - Allow optional heading BEFORE the number but do NOT allow headings that include "act-print"
boundary_v2 = re.compile(
    r"(?m)(?P<heading>[A-Za-z][A-Za-z \-']{0,80}?)?\s*(?P<section>\d{1,4})\.\s*"
)

MAX_REASONABLE_SECTION = 600

matches_v2 = []
for m in boundary_v2.finditer(full_text_v2):
    sec = int(m.group("section"))
    if sec > MAX_REASONABLE_SECTION:
        continue

    # Avoid if heading contains obvious footer junk
    heading = (m.group("heading") or "").strip()
    if "act-print" in heading.lower():
        continue
    if "bdlaws" in heading.lower():
        continue

    # Avoid if preceded by a digit (prevents 1961-like)
    if m.start() > 0 and full_text_v2[m.start()-1].isdigit():
        continue

    # Bracket filter (amendment notes)
    ctx = full_text_v2[max(0, m.start()-25):m.start()+10]
    if "[" in ctx and "]" not in ctx:
        continue

    heading = re.sub(r"\s+", " ", heading)
    matches_v2.append((m.start(), m.end(), sec, heading))

print("Candidate section boundaries found (v2):", len(matches_v2))
print("First 20 candidates (sec, heading) v2:")
for item in matches_v2[:20]:
    print(item[2], "|", item[3])

# Build sections v2; keep the *first* occurrence of a section number as canonical
sections_v2 = {}
for idx, (s_start, s_end, sec, heading) in enumerate(matches_v2):
    next_start = matches_v2[idx + 1][0] if idx + 1 < len(matches_v2) else len(full_text_v2)
    sec_text = full_text_v2[s_end:next_start].strip()

    if sec not in sections_v2:
        sections_v2[sec] = {
            "heading": heading,
            "text": sec_text,
            "start_offset": s_start,
            "end_offset": next_start,
        }

print("\nTotal unique sections parsed (v2):", len(sections_v2))
print("Does section 378 exist (v2)?", 378 in sections_v2)

if 378 in sections_v2:
    print("\n--- Section 378 heading (v2) ---")
    print(sections_v2[378]["heading"])
    print("\n--- Section 378 preview (v2) ---")
    print(sections_v2[378]["text"][:800])
```

```python
import re
import uuid
from typing import List, Dict, Any

from langchain_core.documents import Document  # if this import fails, tell me; we’ll use langchain_core.documents instead.

# 1) Helper: compute page range from <<<PAGE:x>>> markers inside a section body
PAGE_TAG_RE = re.compile(r"<<<PAGE:(\d+)>>>")

def extract_page_range(text: str) -> tuple[int | None, int | None]:
    pages = [int(p) for p in PAGE_TAG_RE.findall(text)]
    if not pages:
        return None, None
    return min(pages), max(pages)

def strip_page_tags(text: str) -> str:
    # Remove tags and clean extra whitespace
    text = PAGE_TAG_RE.sub("", text)
    text = re.sub(r"\n{3,}", "\n\n", text).strip()
    return text

# 2) Chunker: chunk within a section if too long (keeps section metadata)
# NOTE: We chunk at ~900-1200 characters; we’ll tune later after retrieval tests.
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=2000,
    chunk_overlap=300,
    separators=["\n\n", "\n", ". ", " ", ""],
)

ACT_NAME = provenance["act_name"]
ACT_YEAR = provenance["act_year"]
DOC_SHA = provenance["sha256"]

# 3) Convert sections_v2 into Documents ready for vector DB
docs_for_db: List[Document] = []
ids_for_db: List[str] = []

# Sort sections by section number for stable ordering
for sec_num in sorted(sections_v2.keys()):
    if not isinstance(sec_num, int) or sec_num <= 0:
        continue
    sec_obj = sections_v2[sec_num]
    heading = (sec_obj.get("heading") or "").strip()
    raw_text = sec_obj.get("text") or ""

    page_start, page_end = extract_page_range(raw_text)
    clean_text = strip_page_tags(raw_text)

    # If section text is empty after cleaning, skip (should be rare)
    if len(clean_text) < 40:
        continue

    # Split section into chunks (if needed)
    chunks = splitter.split_text(clean_text)

    for idx, chunk_text in enumerate(chunks):
        # Stable deterministic chunk_id:
        # based on sha + act + section + chunk_index (so reruns remain stable if source unchanged)
        chunk_id = f"{DOC_SHA[:12]}::{ACT_NAME}::{sec_num}::{idx}"

        meta: Dict[str, Any] = {
    "law_id": "contract_act_1872",
    "law_name": "Contract Act, 1872",
    "law_short_name": "Contract Act",
    "act_name": ACT_NAME,
    "act_year": ACT_YEAR,
    "section_number": sec_num,
    "section_heading": heading,
    "page_start": page_start,
    "page_end": page_end,
    "source_pdf": provenance["file_name"],
    "source_sha256": DOC_SHA,
}

        docs_for_db.append(Document(page_content=chunk_text, metadata=meta))
        ids_for_db.append(chunk_id)

print("Total chunks prepared for DB:", len(docs_for_db))
print("Sample chunk metadata:", docs_for_db[0].metadata if docs_for_db else None)
print("Sample chunk text (first 300 chars):")
print(docs_for_db[0].page_content[:300] if docs_for_db else "")

# 4) Persist to Chroma (local directory)
PERSIST_DIR = "./chroma_contract_act_1872_v1"
COLLECTION_NAME = "bd_contract_act_1872_en_v1"

from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

model_name = "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
embeddings = HuggingFaceEmbeddings(model_name=model_name)

vectordb = Chroma.from_documents(
    documents=docs_for_db,
    embedding=embeddings,
    ids=ids_for_db,
    collection_name=COLLECTION_NAME,
    persist_directory=PERSIST_DIR,
)

vectordb.persist()
print("\n✅ Chroma persisted.")
print("Persist dir:", PERSIST_DIR)
print("Collection:", COLLECTION_NAME)
```

```python
import re

def rerank_definition(query: str, docs_with_scores):
    q = query.lower()
    wants_definition = any(kw in q for kw in ["define", "definition", "meaning", "what is", "what constitutes", "is said to"])
    if not wants_definition:
        return docs_with_scores  # no rerank

    def_score_re = re.compile(r"\bis said to\b|\bmeans\b|\bdenotes\b|\bwhoever\b", re.IGNORECASE)

    rescored = []
    for doc, dist in docs_with_scores:
        txt = doc.page_content[:600]
        md = doc.metadata
        heading = (md.get("section_heading") or "").lower()

        bonus = 0.0
        # Prefer heading exact match for the term
        if "theft" in q and "theft" == heading.strip():
            bonus += 2.0
        # Prefer definitional language in the first part
        if def_score_re.search(txt):
            bonus += 1.0
        # Penalize punishment sections when definition intent detected
        if "punish" in txt.lower() or "shall be punished" in txt.lower():
            bonus -= 0.8

        # Chroma returns distance-like scores; lower is better.
        # We'll subtract bonus from distance to improve rank.
        rescored.append((doc, dist - bonus, dist, bonus))

    rescored.sort(key=lambda x: x[1])
    return [(d, newdist) for (d, newdist, olddist, bonus) in rescored]

def show_reranked(query: str, k_initial: int = 15, k_final: int = 8):
    print("\n" + "="*95)
    print("QUERY:", query)

    initial = vectordb_v2.similarity_search_with_score(query, k=k_initial)
    reranked = rerank_definition(query, initial)

    print("\nTop results AFTER definitional rerank:")
    for rank, (doc, score) in enumerate(reranked[:k_final], start=1):
        md = doc.metadata
        sec = md.get("section_number")
        heading = md.get("section_heading")
        ps, pe = md.get("page_start"), md.get("page_end")
        snippet = doc.page_content[:220].replace("\n", " ")
        print(f"\n#{rank}  score={score:.4f} | Section {sec} | {heading} | pages {ps}-{pe}")
        print("Snippet:", snippet)

show_reranked("What is theft under Penal Code?")
show_reranked("Definition of theft under Penal Code")
```

```python
def build_evidence_bundle(query: str, k_initial: int = 15, k_final: int = 5):
    # Retrieve
    initial = vectordb_v2.similarity_search_with_score(query, k=k_initial)
    reranked = rerank_definition(query, initial)  # your deterministic reranker
    top = reranked[:k_final]

    bundle = []
    for doc, score in top:
        md = doc.metadata
        bundle.append({
            "score": float(score),
            "act": f"{md.get('act_name')}, {md.get('act_year')}",
            "section": md.get("section_number"),
            "heading": md.get("section_heading"),
            "pages": (md.get("page_start"), md.get("page_end")),
            "source_pdf": md.get("source_pdf"),
            "text": doc.page_content,
        })
    return bundle

bundle = build_evidence_bundle("What is theft under Penal Code?", k_initial=15, k_final=3)
for b in bundle:
    print("\n---")
    print(f"{b['act']} | Section {b['section']} | {b['heading']} | pages {b['pages'][0]}-{b['pages'][1]}")
    print("Score:", b["score"])
    print("Text preview:", b["text"][:350].replace("\n"," "))
```

```python
def format_citation(item: dict) -> str:
    act = item["act"]
    sec = item["section"]
    heading = item.get("heading") or ""
    ps, pe = item.get("pages", (None, None))

    parts = [f"{act} — Section {sec}"]
    if heading.strip():
        parts[-1] += f" ({heading.strip()})"

    if ps is not None and pe is not None:
        if ps == pe:
            parts.append(f"pp. {ps}")
        else:
            parts.append(f"pp. {ps}–{pe}")
    return ", ".join(parts)

def should_refuse(evidence: list, max_score_allowed: float = 20.0) -> tuple[bool, str | None]:
    """
    Conservative refusal policy:
    - refuse if no evidence
    - refuse if the best score is too weak (tune later)
    NOTE: your scores are distance-like; smaller is better.
    """
    if not evidence:
        return True, "No sources retrieved."
    best = evidence[0]["score"]
    if best > max_score_allowed:
        return True, f"Retrieval confidence too low (best_score={best:.2f})."
    return False, None

def build_structured_response(query: str, k_final: int = 3):
    evidence = build_evidence_bundle(query, k_initial=15, k_final=k_final)

    refuse, reason = should_refuse(evidence, max_score_allowed=18.0)

    citations = [format_citation(e) for e in evidence] if not refuse else []

    # For now, no LLM. Return a factual “evidence-first” response:
    if refuse:
        answer_text = ""
    else:
        # Minimal deterministic answer: show definitional section snippet.
        top = evidence[0]
        answer_text = (
            f"According to {top['act']} Section {top['section']} ({top.get('heading','')}), "
            f"the relevant definition is:\n\n"
            f"{top['text'].splitlines()[-1] if len(top['text'].splitlines())>0 else top['text']}"
        )

    return {
        "query": query,
        "answer_text": answer_text,
        "citations": citations,
        "evidence": evidence,
        "refusal_reason": reason,
    }

resp = build_structured_response("What is theft under Penal Code?")
print("Answer:\n", resp["answer_text"])
print("\nCitations:")
for c in resp["citations"]:
    print("-", c)
```

```python
import re
import uuid
from typing import List, Dict, Any

from langchain_core.documents import Document

PAGE_TAG_RE = re.compile(r"<<<PAGE:(\d+)>>>")

def extract_page_range(text: str) -> tuple[int | None, int | None]:
    pages = [int(p) for p in PAGE_TAG_RE.findall(text)]
    if not pages:
        return None, None
    return min(pages), max(pages)

def strip_page_tags(text: str) -> str:
    text = PAGE_TAG_RE.sub("", text)
    text = re.sub(r"\n{3,}", "\n\n", text).strip()
    return text

from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=2000,
    chunk_overlap=300,
    separators=["\n\n", "\n", ". ", " ", ""],
)

ACT_NAME = provenance["act_name"]
ACT_YEAR = provenance["act_year"]
DOC_SHA = provenance["sha256"]

docs_for_db: List[Document] = []
ids_for_db: List[str] = []

for sec_num in sorted(sections_v2.keys()):
    if not isinstance(sec_num, int) or sec_num <= 0:
        continue

    sec_obj = sections_v2[sec_num]
    heading = (sec_obj.get("heading") or "").strip()
    raw_text = sec_obj.get("text") or ""

    page_start, page_end = extract_page_range(raw_text)
    clean_text = strip_page_tags(raw_text)

    if len(clean_text) < 40:
        continue

    chunks = splitter.split_text(clean_text)

    for idx, chunk_text in enumerate(chunks):
        chunk_id = f"{DOC_SHA[:12]}::{ACT_NAME}::{sec_num}::{idx}"

        meta: Dict[str, Any] = {
            "law_id": "contract_act_1872",
            "law_name": "Contract Act, 1872",
            "law_short_name": "Contract Act",
            "act_name": ACT_NAME,
            "act_year": ACT_YEAR,
            "section_number": sec_num,
            "section_heading": heading,
            "page_start": page_start,
            "page_end": page_end,
            "source_pdf": provenance["file_name"],
            "source_sha256": DOC_SHA,
        }

        docs_for_db.append(Document(page_content=chunk_text, metadata=meta))
        ids_for_db.append(chunk_id)

print("Total chunks prepared for DB:", len(docs_for_db))
print("Sample chunk metadata:", docs_for_db[0].metadata if docs_for_db else None)
print("Sample chunk text (first 300 chars):")
print(docs_for_db[0].page_content[:300] if docs_for_db else "")

PERSIST_DIR = "./chroma_contract_act_1872_v1"
COLLECTION_NAME = "bd_contract_act_1872_en_v1"

from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

model_name = "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
embeddings = HuggingFaceEmbeddings(model_name=model_name)

vectordb = Chroma.from_documents(
    documents=docs_for_db,
    embedding=embeddings,
    ids=ids_for_db,
    collection_name=COLLECTION_NAME,
    persist_directory=PERSIST_DIR,
)

vectordb.persist()
print("\n✅ Chroma persisted.")
print("Persist dir:", PERSIST_DIR)
print("Collection:", COLLECTION_NAME)
```
